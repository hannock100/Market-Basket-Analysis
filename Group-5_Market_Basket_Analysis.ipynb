## Answering the Questions

Based on the market basket analysis performed using the Apriori algorithm with a support threshold of 10% and the generated association rules, here are the answers to your questions:

**1. Top Seller - Which of the four companies' back cover has sold the most?**

To answer this, we'll look at the `support` of individual mobile covers (M1, M2, M3, M4) in the `frequent_itemsets` DataFrame. The item with the highest support is the top seller among the back covers.

```python
mobile_covers = ['M1', 'M2', 'M3', 'M4']
top_selling_cover = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: list(x)[0] in mobile_covers and len(x) == 1)].sort_values(by='support', ascending=False).iloc[0]

print(f"The top selling back cover is: {list(top_selling_cover['itemsets'])[0]} with a support of {top_selling_cover['support']:.2f}")
```

**2. Not Eligible for strategy - Which of the following tempered glass are not eligible for the strategy when you take one product at a time?**

Products not eligible for a strategy when taken one at a time are those with a support below the 10% threshold. We need to check the support of individual tempered glasses (T1, T2, T3).

```python
tempered_glasses = ['T1', 'T2', 'T3']
ineligible_glasses = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: list(x)[0] in tempered_glasses and len(x) == 1 and x not in frequent_itemsets[frequent_itemsets['support'] >= 0.1]['itemsets'])]

if ineligible_glasses.empty:
    print("All tempered glasses are eligible for the strategy when taken one at a time based on the 10% support threshold.")
else:
    print("The following tempered glasses are not eligible for the strategy when you take one product at a time:")
    for index, row in ineligible_glasses.iterrows():
        print(f"- {list(row['itemsets'])[0]} with a support of {row['support']:.2f}")
```

**3. Top Pairs - When you take two products at a time, which of these pairs have sold the most?**

To find the top selling pairs, we look at the `support` of itemsets with two items in the `frequent_itemsets` DataFrame. The pair with the highest support has sold the most.

```python
top_selling_pair = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 2)].sort_values(by='support', ascending=False).iloc[0]

print(f"The top selling pair of products is: {list(top_selling_pair['itemsets'])} with a support of {top_selling_pair['support']:.2f}")
```

**4. Eligible for Strategy - Which of the following pairs of products are eligible for strategy according to the threshold?**

Pairs of products eligible for strategy are those with a support of 10% or higher. These are already included in our `frequent_itemsets` DataFrame for itemsets with a length of 2.

```python
eligible_pairs = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 2 and x in frequent_itemsets[frequent_itemsets['support'] >= 0.1]['itemsets'])]

if eligible_pairs.empty:
    print("No pairs of products are eligible for strategy according to the 10% support threshold.")
else:
    print("The following pairs of products are eligible for strategy according to the 10% support threshold:")
    for index, row in eligible_pairs.iterrows():
        print(f"- {list(row['itemsets'])} with a support of {row['support']:.2f}")
```

**5. Profitability - What is the probability that someone will buy the mobile cover M4 given that he/she has purchased the tempered glass T3?**

This probability is represented by the **confidence** of the rule {T3} -> {M4}. We can find this in the `rules` DataFrame.

```python
probability_m4_given_t3 = rules[(rules['antecedents'].apply(lambda x: 'T3' in x)) &
                                (rules['consequents'].apply(lambda x: 'M4' in x))]['confidence'].iloc[0]

print(f"The probability that someone will buy mobile cover M4 given they purchased tempered glass T3 is: {probability_m4_given_t3:.2f}")
```

**6. Cross-Selling Strategy - If someone has purchased the mobile cover M1, which tempered glass should you recommend?**

To determine which tempered glass to recommend if a customer buys M1, we should look at the association rules where M1 is the antecedent and a tempered glass is the consequent. We would recommend the tempered glass with the highest **confidence** in these rules.

```python
m1_rules = rules[(rules['antecedents'].apply(lambda x: 'M1' in x)) &
                 (rules['consequents'].apply(lambda x: list(x)[0] in tempered_glasses))]

if m1_rules.empty:
    print("There are no association rules recommending a tempered glass given the purchase of M1.")
else:
    recommended_glass = m1_rules.sort_values(by='confidence', ascending=False).iloc[0]
    print(f"If someone has purchased the mobile cover M1, you should recommend the tempered glass: {list(recommended_glass['consequents'])[0]} with a confidence of {recommended_glass['confidence']:.2f}")
```
